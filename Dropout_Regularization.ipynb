```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv("sonar_dataset.csv",header=None)
df.sample(5)
```

# Explore the Date

```python
df.shape
```

```python
df.isnull().sum()
```

```python
df[60].value_counts()
```

```python
x = df.drop(60,axis="columns")
y = df[60]
y.head()
```

```python
y = pd.get_dummies(y,drop_first=True)
y.sample(5)
```

```python
y.value_counts()
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test = train_test_split(x,y,test_size=0.2,random_state=42)
```

```python
x_train.shape,x_test.shape
```

```python
import tensorflow as tf
from tensorflow import keras
```

```python
model = keras.Sequential([
    keras.layers.Dense(60,input_dim=60,activation="relu"),
    keras.layers.Dense(30,activation="relu"),
    keras.layers.Dense(15,activation="relu"),
    keras.layers.Dense(1,activation="sigmoid")
])
```

```python
model.compile(loss = "binary_crossentropy",optimizer="adam",metrics=["accuracy"])
```

```python
model.fit(x_train,y_train,epochs = 100,batch_size=8)
```

# see the accuracy of the model is 1

# lets test the acuracy on test_set

```python
model.evaluate(x_test,y_test)
```

# Now the accuracy of test is less then training  this seem like model is overfitted

```python
y_pred = model.predict(x_test).reshape(-1)
print(y_pred[:10])

# round the values to nearest integer 0 or 1 
y_pred = np.round(y_pred)
print(y_pred[:10])
```

```python
y_test
```

```python
from sklearn.metrics import confusion_matrix , classification_report
print(classification_report(y_test,y_pred))
```

# now just try a model with drop out layer

```python
modeld = keras.Sequential([
    keras.layers.Dense(60,input_dim=60,activation="relu"),
    keras.layers.Dropout(0.5),
    keras.layers.Dense(30,activation="relu"),
    keras.layers.Dropout(0.5),
    keras.layers.Dense(15,activation="relu"),
    keras.layers.Dropout(0.5),
    keras.layers.Dense(1,activation="sigmoid")
])
modeld.compile(loss = "binary_crossentropy",optimizer="adam",metrics=["accuracy"])
modeld.fit(x_train,y_train,epochs = 100,batch_size=8)
```

# with dropout layer  accuracy of training set is 90%

```python
modeld.evaluate(x_test,y_test)
```

# See the accuracy of test and training set is same 

```python
y_pred = modeld.predict(x_test).reshape(-1)
y_pred = np.round(y_pred)
```

```python
from sklearn.metrics import confusion_matrix , classification_report
print(classification_report(y_test,y_pred))
```
